```python
import torch
from diffusers import StableVideoDiffusionPipeline
from PIL import Image
import imageio
import numpy as np

# Load model
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableVideoDiffusionPipeline.from_pretrained(
    "stabilityai/stable-video-diffusion-img2vid-xt",
    torch_dtype=torch.float16,
    variant="fp16",
).to(device)
pipe.enable_model_cpu_offload()

# Load and resize input image
input_image = Image.open("/content/drive/MyDrive/AI/Video_Model_Stable_Diffusion/Capture d'écran 2025-05-16 002901.png").convert("RGB")
input_image = input_image.resize((1024, 576))  # Model's required resolution

# Generate frames
generator = torch.manual_seed(42)
frames = pipe(
    input_image,
    num_frames=25,
    fps=7,
    generator=generator,
    decode_chunk_size=5,
).frames[0]

# Save as GIF
gif_path = "output.gif"
imageio.mimsave(
    gif_path,
    [np.array(frame) for frame in frames],
    duration=1000 // 7,  # 7 FPS = ~142ms per frame
    loop=0,
)

print(f"GIF saved to {gif_path}")
```

```python
import flask
from flask import Flask, request, send_file
import torch
from diffusers import StableVideoDiffusionPipeline
from PIL import Image
import imageio
import numpy as np
import io
import uuid
import subprocess
from flask_cors import CORS


app = Flask(__name__)
CORS(app)

# Install dependencies
def install_dependencies():
    subprocess.run(['pip', 'install', 'pyngrok', 'flask', 'torch', 'diffusers', 'accelerate', 'imageio'], check=True)

install_dependencies()

from pyngrok import ngrok

# Initialize pipeline
pipe = None

def initialize_pipeline():
    global pipe
    pipe = StableVideoDiffusionPipeline.from_pretrained(
        "stabilityai/stable-video-diffusion-img2vid-xt",
        torch_dtype=torch.float16,
        variant="fp16",
    )
    pipe.enable_model_cpu_offload()

initialize_pipeline()

@app.route('/generate-video', methods=['POST'])
def generate_video():
    if 'image' not in request.files:
        return "No image uploaded", 400

    file = request.files['image']
    if file.filename == '':
        return "No selected file", 400

    # Process image
    input_image = Image.open(file.stream).convert("RGB")
    input_image = input_image.resize((512, 512))  # Model's required resolution

    # Generate frames
    generator = torch.manual_seed(42)
    frames = pipe(
        input_image,
        num_frames=25,
        fps=7,
        generator=generator,
        decode_chunk_size=5,
    ).frames[0]

    # Save to GIF in memory
    gif_bytes = io.BytesIO()
    imageio.mimsave(
        gif_bytes,
        [np.array(frame) for frame in frames],
        format='GIF',
        duration=1000 // 7,
        loop=0,
    )
    gif_bytes.seek(0)

    return send_file(gif_bytes, mimetype='image/gif')

if __name__ == '__main__':
    # Set up ngrok
    ngrok.set_auth_token("2wi0e1dFfnrobkC2XL8PORw1mzz_5vWecgHecPbYERm8TPVsX")
    public_url = ngrok.connect(5050).public_url
    print(f" * Running on: {public_url}")

    app.run(host='0.0.0.0', port=5050)
```
